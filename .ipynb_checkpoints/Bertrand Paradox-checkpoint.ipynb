### Bertrand's Paradox

__Q__: Suppose a chord on the unit circle is chosen at random. What is the probability that the length of the chord is greater than $\sqrt{3}$?

Observe that an equilateral triangle inscribed on the unit circle has side length $\sqrt{3}$ which gives us an easy way
to *check* whether a chord has length greater than $\sqrt{3}$ by comparison an appropriate equilateral triangle.

Consider [three methods]( https://en.wikipedia.org/wiki/Bertrand_paradox_%28probability%29#Bertrand's_formulation_of_the_problem ) of choosing a random chord:

1. __Choosing two random endpoints.__ We choose two endpoints $p$ and $q$ randomly on the circumference of the circle that generates a unique chord. It should be noted that every chord on the circle corresponds to two endpoints $p$ and $q$. By symmetry, the resulting length of the chord relies purely on the choosing of $q$. By constructing an equilateral triangle with a vertex on $p$, we see that $\frac{1}{3}$ choices of $q$ (contained in the bottom base of the equilateral triangle) result to chords longer than the sides of the equilateral triangle emanating from $p$.


2. __Choosing a random radius.__ The set of chords on the circle can be put into 1-1 correspondence with a radial vector that bisects it. Thus a random choice of chord can be done by a random choice of a directed radial segment of length $r$. By symmetry, this is a choice of $r \in [0,1)$. Construct the chord that is perpendicular to the chosen radius. Compare with an equilateral triangle whose side is perpendicular to the choice of random radius and note that the circle inscribed inside the inscribed triangle has radius $\frac{1}{2}$.


3. __Choosing a random midpoint.__ The set of chords on the unit circle can be corresponds to the set of points on the circle, i.e. the midpoint $m$ of each chord. So a chord can be chosen randomly by randomly choosing a point $m$ inside the circle. By comparison with an equilateral triangle with side parallel to the generated chord, we see that midpoints chosen inside the circle inscribed in the equilateral triangle (with radius $\frac{1}{2}$) generates a chord with length greater than $\sqrt{3}$. This happens $\pi\left(\frac{1}{2}\right)^2 / \pi = \frac{1}{4}$ of the time assuming a uniform choice of points on the circle.

__Paradox__: The paradox is that the three methods give a probability of $\frac{1}{3}, \frac{1}{2}$, and $\frac{1}{4}$, respectively. This can be resolved by the fundamental insight that a measure of a probability of an event (a set of outcomes) is tied to the *process* that generates the event. 

In probability theory, this process, called a *random experiment*, is not defined. What is defined is a sample space $\Omega$ which is simply a set containing all outcomes of the random experiment, and a probability measure $\mathbb{P}$ that gives a measure of likelihood $\mathbb{P}(A) \in [0,1]$ that an outcome $\omega \in \Omega$ is in $A$. In the three methods above, each has its own sample spaces and probability measure, so that there is no real paradox. 

For example, in method 1, 
we make a random choice on the arclength $s \in (0, 2\pi)$ that, despite being a uniform choice on the interval, results to a nonuniform distribution of chords on the circle (as shown by the nonuniform distribution of midpoints, by density). This is in contrast with method 3 which, by construction, results to a uniform distribution of midpoints on the circle (by density).

### Experiments

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# iterate over the first fixed endpoint
mx = np.empty(10000)
my = np.empty(10000)
for k in range(10000):
    t = np.random.random()*2*np.pi   # random rotation angle = random choice of first endpoint
    s = np.random.random()*2*np.pi   # random choice of second endpt from (1,0)
    x = (np.cos(s) + 1)/2  
    y = np.sin(s)/2
    x, y = np.cos(t)*x - np.sin(t)*y, np.sin(t)*x + np.cos(t)*y  # rotate from (1,0) to correct orientation
    mx[k] = x
    my[k] = y

plt.scatter(mx, my, alpha=0.1, marker='.', color='b')
plt.axis('scaled');
```

Random endpoint method

```python
x = np.empty(10000)
y = np.empty(10000)
for i in range(10000):
    r = np.random.random()
    theta = np.random.random()*2*np.pi
    x[i] = r*np.cos(theta)
    y[i] = r*np.sin(theta)

plt.scatter(x, y, alpha=0.1, marker='.', color='b')
plt.axis('scaled');
```

Random radius method

```python
x = np.empty(10000)
y = np.empty(10000)
k = 0
while k < 10000:
    i = np.random.choice([1, -1])*np.random.random()
    j = np.random.choice([1, -1])*np.random.random()
    if i*i + j*j > 1:
        continue
    x[k] = i
    y[k] = j
    k += 1
        
plt.scatter(x, y, alpha=0.1, marker='.', color='b')
plt.axis('scaled');
```
